```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
heart_disease = fetch_ucirepo(id=45)

# data (as pandas dataframes)
X = heart_disease.data.features
y = heart_disease.data.targets

# metadata
print(heart_disease.metadata)

# variable information
print(heart_disease.variables)
```

```python
X.head()
```

```python
import numpy as np
# Count the number of unique classes in the target variable 'y'
num_classes = len(np.unique(y))

print(f'Number of classes: {num_classes}')
```

```python
import pandas as pd

# Assuming 'X' is your dataset as a Pandas DataFrame
# You can check for NaN values in the entire DataFrame
if X.isnull().values.any():
    print("There are NaN values in the dataset.")
else:
    print("No NaN values found in the dataset.")
```

```python
import pandas as pd

X = X.fillna(X.mean())

# Check if there are any remaining NaN values
if X.isnull().values.any():
    print("Some NaN values remain in the dataset.")
else:
    print("All NaN values have been imputed with the mean.")
```

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# Create a MinMaxScaler instance
scaler = MinMaxScaler()

# Fit and transform the dataset using the scaler
X_normalized = scaler.fit_transform(X)

# Convert the normalized data back to a DataFrame (optional)
X_normalized_df = pd.DataFrame(X_normalized, columns=X.columns)
X_normalized_df
```

```python
from sklearn.model_selection import train_test_split

# Spliting the data into a training set and a testing set (e.g., 80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report

# Assuming 'X_train' and 'y_train' are your training data
# Assuming 'X_test' and 'y_test' are your testing data

# Train a Logistic Regression model
logistic_model = LogisticRegression(max_iter=1000, random_state=42)
logistic_model.fit(X_train, y_train)

# Make predictions on the test data using Logistic Regression
logistic_predictions = logistic_model.predict(X_test)

# Train a Gradient Boosting model
gradient_boosting_model = GradientBoostingClassifier(n_estimators=100, random_state=42)
gradient_boosting_model.fit(X_train, y_train)

# Make predictions on the test data using Gradient Boosting
gradient_boosting_predictions = gradient_boosting_model.predict(X_test)

```

```python

# Calculate accuracy and print a classification report for Logistic Regression
logistic_accuracy = accuracy_score(y_test, logistic_predictions)
logistic_classification_report = classification_report(y_test, logistic_predictions)

# Calculate accuracy and print a classification report for Gradient Boosting
gradient_boosting_accuracy = accuracy_score(y_test, gradient_boosting_predictions)
gradient_boosting_classification_report = classification_report(y_test, gradient_boosting_predictions)

# Compare the accuracies and classification reports of both models
print(f"Logistic Regression Accuracy: {logistic_accuracy:.2f}")
print(f"Logistic Regression Classification Report:\n{logistic_classification_report}")
print(f"Gradient Boosting Accuracy: {gradient_boosting_accuracy:.2f}")
print(f"Gradient Boosting Classification Report:\n{gradient_boosting_classification_report}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Set the style for Seaborn
sns.set(font_scale=1.2)
plt.figure(figsize=(12, 6))

# Generate confusion matrices for both models
logistic_confusion_matrix = confusion_matrix(y_test, logistic_predictions)
gradient_boosting_confusion_matrix = confusion_matrix(y_test, gradient_boosting_predictions)

# Get unique class labels for your multiclass problem
unique_labels = np.unique(y_test)

# Plot the confusion matrix for Logistic Regression
plt.subplot(1, 2, 1)
sns.heatmap(logistic_confusion_matrix, annot=True, fmt="d", cmap="Blues",
            xticklabels=unique_labels, yticklabels=unique_labels, cbar=False)
plt.title("Logistic Regression Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")

# Plot the confusion matrix for Gradient Boosting
plt.subplot(1, 2, 2)
sns.heatmap(gradient_boosting_confusion_matrix, annot=True, fmt="d", cmap="Blues",
            xticklabels=unique_labels, yticklabels=unique_labels, cbar=False)
plt.title("Gradient Boosting Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")

plt.tight_layout()
plt.show()
```
